```python
# for döngüsü örneği: bir sınıftaki öğrencilerin notlarının ortalamasını hesaplama
notlar = [85, 92, 78, 90, 88]

toplam = 0
for sayi in notlar:
    toplam += sayi     # toplam = toplam + sayi

ortalama = toplam / len(notlar)

print("Birinci Print -> Not Ortalaması: ", ortalama)

print(f"İkinci Print -> Not ortalaması : {ortalama}")
```

```python
# while döngüsü örneği: 0'dan 5'e kadar olan sayıları yazdırılması işlemidir.

i = 5
while (i >= 0) :
  print(i)
  i -= 1
```

```python
# kullanıcıdan bilgi alınması
while True:
  try:
    gelen_bilgi = int(input("Bir sayı gir: "))
    break
  except ValueError:
    print("Lütfen bir sayı gir:")
i = 0
while i <= gelen_bilgi:
  print(i)
  i += 1
```

```python
# while döngüsü örneği: kullanıcı bakiye bilgisinin sorgulanması

toplam_bakiye = 1000

while toplam_bakiye > 0:
  try:
    cekilen_tutar = int(input("Çekmek istediğiniz tutarı girin: "))
    if cekilen_tutar > toplam_bakiye:
      print("Bakiyeniz yetersiz.")
      bilgi = input("İşleme devam etmek istiyor musunuz? (E/H): ").lower()
      if bilgi == "h":
        print("İşlem sonlandırıldı.")
        break
      else:
        print("İşlem devam ediyor...")
        continue
    else:
      toplam_bakiye -= cekilen_tutar
      print(f"Kalan bakiye: {toplam_bakiye}")
      if toplam_bakiye == 0:
        print("Bakiye tükendi.")
  except ValueError:
      print("Lütfen geçerli bir sayı girin!")
```

```python
# Önceden tanımlalan sesli harf listesinde girilen metnin karşılaştırılması yapılarak sesli ve sessiz harf kontrolünün yapımı
sesli_harfler = 'aeıioöuüAEIİOÖUÜ'

kelime = input("Bir kelime girin: ") # Kullanıcıdan bir kelime al

# Sesli ve sessiz harf sayısını hesapla
sesli_harf_sayisi = 0
sessiz_harf_sayisi = 0

'''
isalpha() ile bir stringde yalnızca harf olup olmadığını kontrol etmek için kullanılır.

Aşağıdaki kod blogunda geçen "isalpha" harf dışı bir karakter ile karşılaşınca "false" döndürmektedir.
'''

for harf in kelime:
    if harf.isalpha():  # Sadece harfleri kontrol et
        if harf in sesli_harfler:
            sesli_harf_sayisi += 1
        else:
            sessiz_harf_sayisi += 1

# Sonuçları ekrana yazdır
print(kelime, "kelimesinde" , sesli_harf_sayisi , "adet sesli harf ve", sessiz_harf_sayisi, "sessiz harf vardır.")
```
